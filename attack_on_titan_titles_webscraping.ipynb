## Using Webscraping to Extract Title Name from wiki


```python
html_data=requests.get('https://zh.wikipedia.org/wiki/進擊的巨人#各話列表').text
```

Parse the html data using `beautiful_soup`.


```python
soup = BeautifulSoup(html_data,'html5lib')
```

```python
table=soup.find_all('table',class_="wikitable")
```

```python
my_table=table[11]
```

```python
rows=my_table.find("tbody").find_all("tr")
```

```python
del rows[0:2]
```

```python
rows[0].find_all('td')
```

```python
title=pd.DataFrame(columns=['話數','中文標題',])
```

```python
for row in rows:
    if len(row)>3:
        col = row.find_all('td')
        number =col[0].text.replace("#",'')
        name=col[2].text
        title=title.append({'話數':number,'中文標題':name},ignore_index=True)
    else:
        pass
```

```python
title
```

```python
for i in range(0,76):
    print(title.iloc[i,0],title.iloc[i,1])
```
