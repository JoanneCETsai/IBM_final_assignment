```python
#引入模組

import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import urllib3
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split

%matplotlib inline
```

```python
# 1.Data Generation
x = np.random.uniform(-1.5, 1, size=1000)
X = x.reshape(-1, 1)
#X
```

```python
y=3*x**3 + 2*x**2-3*x+1+np.random.normal(0, 0.5, size=1000)
y=y.reshape(-1,1)
#y
```

```python
plt.scatter(X,y)
```

```python
# 2.Data Preprocessing

poly1= PolynomialFeatures(degree=1)
poly2= PolynomialFeatures(degree=2)
poly3= PolynomialFeatures(degree=3)
poly4= PolynomialFeatures(degree=4)
```

```python
# 3. Model Construction
```

```python
poly1.fit(X)
poly2.fit(X)
poly3.fit(X)
poly4.fit(X)
```

```python
X1=poly1.transform(X)
X2=poly2.transform(X)
X3=poly3.transform(X)
X4=poly4.transform(X)
X1.shape
```

```python
lin_reg1 = LinearRegression()
lin_reg1.fit(X1, y)
y_predict1 = lin_reg1.predict(X1)

lin_reg2 = LinearRegression()
lin_reg2.fit(X2, y)
y_predict2 = lin_reg2.predict(X2)

lin_reg3 = LinearRegression()
lin_reg3.fit(X3, y)
y_predict3 = lin_reg3.predict(X3)

lin_reg4 = LinearRegression()
lin_reg4.fit(X4, y)
y_predict4 = lin_reg4.predict(X4)
```

```python
plt.scatter(x, y)
plt.plot(np.sort(x), y_predict1[np.argsort(x)],color='r',label='degree=1')
plt.legend()
plt.show()
```

```python
plt.scatter(x, y)
plt.plot(np.sort(x), y_predict2[np.argsort(x)], color='r',label='degree=2')
plt.legend()
plt.show()
```

```python
plt.scatter(x, y)
plt.plot(np.sort(x), y_predict3[np.argsort(x)], color='r',label='degree=3')
plt.legend()
plt.show()
```
